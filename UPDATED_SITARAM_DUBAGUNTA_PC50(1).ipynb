## (a) Five-number summary of numeric attributes

```python

import pandas as pd
import numpy as np
# Load the cereals dataset
df = pd.read_csv("datasets/Cereals.csv")

# Compute five-number summary before any preprocessing
summary_before = df.describe().loc[['min', '25%', '50%', '75%', 'max']]
summary_before
```

## (b) Replace missing values (-1) with mean

```python
import numpy as np

# Convert '?' to NaN directly
df.replace('?', np.nan, inplace=True)

# Ensure -1 is also treated as NaN
df.replace(-1, np.nan, inplace=True)



# Fill NaN values with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)

df.head()
```

## (c) Five-number summary after treating missing values

```python

# Compute five-number summary after treating missing values
summary_after_missing = df.describe().loc[['min', '25%', '50%', '75%', 'max']]
summary_after_missing
```

## (d) Replace noisy data (outliers) with median

```python

# Identify and replace noisy data (outliers) using median
for col in df.select_dtypes(include=['number']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].apply(lambda x: df[col].median() if x < lower_bound or x > upper_bound else x)
df.head()
```

## (e) Five-number summary after treating noisy values

```python

# Compute five-number summary after treating noisy values
summary_after_noisy = df.describe().loc[['min', '25%', '50%', '75%', 'max']]
summary_after_noisy
```
