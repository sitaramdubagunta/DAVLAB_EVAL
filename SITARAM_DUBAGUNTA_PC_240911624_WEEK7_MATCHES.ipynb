7. Visualize the number of wins of each team for the venue of MA Chidambaram Stadium, 
Chepauk 

```python
new_dataset = df[df['venue'] == 'MA Chidambaram Stadium, Chepauk']
new_dataset
```

```python
plt.figure(figsize=(10,10))
team_counts = new_dataset['winner'].value_counts()
plt.pie(team_counts, labels=team_counts.index ,  startangle=140)
plt.title('Distribution of IPL Winners')
plt.show()
```
